```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('insurance.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python

data.isnull().sum()
```

```python
data.sex.value_counts()
```

```python
data.region.value_counts()
```

```python
encoder = LabelEncoder()
```

```python
labels = encoder.fit_transform(data.sex)
```

```python
data['sex'] = labels
```

```python
data.head()
```

```python

labels = encoder.fit_transform(data.region)
```

```python
data['region'] = labels
```

```python

data.head()
```

```python
labels = encoder.fit_transform(data.smoker)
```

```python
data['smoker'] = labels
```

```python
data.head()
```

```python
X = data.drop(columns='charges',axis=1)
Y = data['charges']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
testing_data_prediction = model.predict(X_test)
```

```python
score = metrics.r2_score(Y_test,testing_data_prediction)
score
```

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")
input_data = (19,0,27.9,0,1,3)
input_data_array = np.asarray(input_data)

#reshaping the data so that it works for only one instance at a time
input_data_reshaped = input_data_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print('Predicted Medical Insurance Cost : ',str(prediction))
```
